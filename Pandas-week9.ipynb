```python
#Pivot Table
df = pd.DataFrame({
    "Department": ["IT", "IT", "HR", "HR", "Finance"],
    "Gender": ["M", "F", "M", "F", "M"],
    "Salary": [50000, 55000, 30000, 35000, 40000]
})
print("Average salary per department")
avg=df.pivot_table(values="Salary",
                   index="Department",
                   aggfunc="mean")
print(avg)
print()

print("Average salary per department + gender (pivot table)")

avg1=df.pivot_table(values="Salary",
                    index=["Department","Gender"],
                    aggfunc="mean")
print(avg1)
print()

print("Pivot table showing sum and count of salary per department")
final=df.pivot_table(values="Salary",
                     index="Department",
                     aggfunc=["sum","count"])
print(final)
```

```python
import pandas as pd
df=pd.DataFrame({"Department":["HR","IT","Finance","HR"],
                 "Employee":["Arun","Neeva","Alan","Meena"],
                 "Salary":[25000,30000,19000,42000],
                 "Experience":[2,3,1,4]})
print("Group by Department and calculate Average Salary, Total Experience and Number of employees")
grouping=df.groupby("Department").agg({
    "Salary":"mean",
    "Experience":"sum",
    "Employee":"count"})
print(grouping)
print()

print("Sort the grouped result by Average Salary, descending")
grp_sorted=grouping.sort_values("Salary",ascending=False)
print(grp_sorted)
print()

print("Reset the index of the final grouped table")
grp_sorted.reset_index()
```

```python
import pandas as pd
Employee_info=pd.DataFrame({"Employee_id":[1,2,3,4],
                            "Name":["Neeva","Arun","Maya","Hena"],
                            "Department":["HR","IT","HR","Finance"]})
Salary_info=pd.DataFrame({"Employee_id":[1,2,4,3],
                           "Department":["HR","IT","HR","IT"],
                           "Salary":[20000,32000,16000,47000]})
print("Merge the two DataFrames on EmployeeID and Department")
a=pd.merge(Employee_info,Salary_info, on=["Employee_id","Department"],how="inner")
print(a)
print()

print("Inner merge")
b=pd.merge(Employee_info,Salary_info,on=["Employee_id","Department"],how="outer")
print(b)
print()

print("Left merge")
d=pd.merge(Employee_info,Salary_info,on=["Employee_id","Department"],how="left")
print(d)
print()

e = pd.merge(Employee_info, Salary_info,
             on=["Employee_id", "Department"],
             how="inner",
             suffixes=("_Left", "_Right"))
print(e)
```

```python
import pandas as pd
df=pd.DataFrame({"Name":["Neeva","Arun","Maya","Hena"],
                 "Maths":[90,21,78,97],
                 "Science":[17,56,39,80],
                 "English":[67,47,43,60]})
print("Convert only the Name column to uppercase (use correct function)")
a=df["Name"].map(str.upper)
print(a)
print()

print("Add a new column Total = sum of all subjects using apply(row-wise)")
df["Total"]=df.apply(lambda row: row["Maths"]+row["Science"]+row["English"], axis=1)
print(df)

def column_grade(Total):
    if Total>=240:
        return 'A'
    elif Total>=180:
        return 'B'
    elif Total>=120:
        return 'C'
    else:
        return 'D'
df["Grade"]=df["Total"].apply(column_grade)
print(df)
```
