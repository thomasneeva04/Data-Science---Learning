```python
#iteration
import numpy as np
arr=np.array([[1,2,3,4],[2,4,7,8]])
for a in arr:
    print(a) # this will give result in the array format only

for x in arr:
    for y in x:
        print(y) #this will give the array with each elements
        
print() # just for space

# to make this above step easier, which is to make lesser n for loops we can write
for a in np.nditer(arr):
    print(a)
```

```python
#filtering
import numpy as np
arr=np.array([30,31,32,33,34,35])
new_arr=[]
for element in arr:
    if element>31:
        new_arr.append(True)
    else:
        new_arr.append(False)

a=arr[new_arr]
print(arr)
print(a)
```

```python
import numpy as np
arr=np.random.randint(1,101, size=20)

new=arr.reshape(4,5)
print(new)
print()

rows=new[:2,:] #first 2 rows
print(rows)
print()

cols=new[:,-2:]
print(cols)
print()

new2=np.vstack((new,np.array([999,999,999,999,999])))
print(new2)
print()

addedrow=np.array_split(new2,2, axis=1)
print(addedrow)
```

```python
#Matrix multiplication
A=np.array([[1,2,3],
            [4,5,6]])
            
B=np.array([[10,20],
            [30,40],
            [50,60]])
C=A@B
print("Matrix multiplication:")
print(C)
print()

trans=C.T
print("Transpose: ")
print(trans)
print()

m=trans.mean()
print("Mean: ",m)
print()

square=trans**2
print("squared: ")
print(square)
```

```python
import numpy as np
arr=np.random.randint(0,101,size=(5,5)) # 5*5 matrix generated
print(arr)
print()

print("Avg marks of each student: ")
avg=arr.mean(axis=1)
print(avg)
print()

print("Avg marks in each subject: ")
avg_marks=arr.mean(axis=0)
print(avg_marks)
print()

print("Overall average: ")
all_avg=arr.mean()
print(all_avg)
print()

print("Replacing all values below  40 with 0: ")
arr[arr<40]=0
print(arr)
print()

print("Total marks added: ")
total=arr.sum(axis=1)
total.reshape(5*1)
final=np.hstack((arr,total.reshape(5,1)))
print(final)
print()

print("1-D array: ")
print(arr.flatten())
```
