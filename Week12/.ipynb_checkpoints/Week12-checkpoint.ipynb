##### Finding Standard Error

```python
import numpy as np
sd=np.std(df['Age'],ddof=1)  #ddof means: Delta Degrees Of Freedom
n=len(df['Age'])
SE=sd/np.sqrt(n)
SE.round(3)
```

##### Finding sampling distributions of Sample Mean

```python
samples=[]
for i in range(1000):
    sample=df['Age'].sample()
    samples.append(sample.mean())

samples
```

### T- TEST

##### One- Sample t-test

```python
from scipy.stats import ttest_1samp

stat, p = ttest_1samp(df['Wait_Time_Minutes'], 10)

print(stat, p)
```

##### Two- Sample Independent t-test

```python
from scipy.stats import ttest_ind

stat, p = ttest_ind(group1, group2, equal_var=False)

#Why equal_var=False?
#Because we dont assume equal variances = Welchs t-test

print(stat, p)
```

##### Paired t-test

```python
from scipy.stats import ttest_rel

stat, p = ttest_rel(before, after)
print(stat, p)
```

#### CHI-SQUARE TEST FOR INDEPENDENCE

```python
from scipy.stats import chi2_contingency

table = pd.crosstab(df['Gender'], df['Survived'])
chi2, p, dof, expected = chi2_contingency(table)

print(chi2, p)
```

#### CHI-SQUARE GOODNESS-OF-FIT TEST

```python
from scipy.stats import chisquare

observed = [50, 30, 20]
expected = [40, 40, 20]

chi2, p = chisquare(observed, expected)
print(chi2, p)
```

#### ONEWAY- ANOVA

```python
from scipy.stats import f_oneway

stat, p = f_oneway(groupA, groupB, groupC)
print(stat, p)
```

#### POST-HOC ANOVA

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

tukey = pairwise_tukeyhsd(
    endog=df['Score'],
    groups=df['Teaching_Method'],
    alpha=0.05
)

print(tukey)
```
