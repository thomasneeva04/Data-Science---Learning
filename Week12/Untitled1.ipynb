```python
import pandas as pd

df=pd.read_csv('anova_dataset (1).csv')
df.head()
```

```python
df['Method'].unique()
```

```python
from scipy.stats import f_oneway

grp1= df[df['Method']=='Lecture']['Marks'].dropna()
grp2=df[df['Method']=='Interactive']['Marks'].dropna()
grp3=df[df['Method']=='Blended']['Marks'].dropna()


stats, p= f_oneway(grp1,grp2,grp3)
p
```

```python
df=pd.read_csv('ttest_dataset.csv')
df.head()
```

```python
df['Group'].unique()
```

```python
from scipy.stats import ttest_ind

grp1=df[df['Group']=='Online']['Score'].dropna()
grp2=df[df['Group']=='Offline']['Score'].dropna()

stats,p= ttest_ind(grp1,grp2)
p
```

```python
df=pd.read_csv('chi_dataset.csv')
df.head()
```

```python
from scipy.stats import chi2_contingency

table=pd.crosstab(df['Gender'],df['Preference'])

chi2,p,dof,expected=chi2_contingency(table)
p.round(3)
```
