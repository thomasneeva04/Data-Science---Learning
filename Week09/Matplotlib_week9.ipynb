```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

product_ids=np.random.randint(1001, 1010, size=100)
quantities=np.random.randint(1,5, size=100)
prices=np.random.uniform(10,100,size=100).round(2)
dates=np.random.choice(pd.date_range("2025-01-01","2025-01-10"),size=100)

sales_df=pd.DataFrame({"ProductID":product_ids,
                        "Quantity":quantities,
                        "Price":prices,
                        "Date":dates})
print(sales_df.head(5))
print()

print("Total Sale column: Quantity  Price")
sales_df["total_sale"]=sales_df["Quantity"]*sales_df["Price"]
print(sales_df)
print()

print("Randomly set 5 Price values in sales_df to NaN")
missing_values=np.random.choice(sales_df.index, size=5, replace=False)
sales_df.loc[missing_values, "Price"]=np.nan
print(sales_df.head())
print()

print("Fill NaN with mean price")
sales_df["Price"]=sales_df["Price"].fillna(sales_df["Price"].mean())
print(sales_df.head())
print()

print("Group by ProductID and calculate: Total quantity sold")
print(sales_df.groupby("ProductID")["Quantity"].sum())
print()
print("Total Revenue:")
sales_df["Revenue"]=sales_df["Quantity"]*sales_df["Price"]
total_Revenue=sales_df.groupby("ProductID")["Revenue"].sum()
print(total_Revenue)
print()

plt.subplot(1,2,1)
plt.bar(total_Revenue.index,total_Revenue.values, color="Orange")
plt.title("Revenue per ProductID")
plt.xlabel("ProductID")
plt.ylabel("Revenue")

plt.subplot(1,2,2)
plt.hist(sales_df["Revenue"], bins=10, color="purple")
plt.title("Total Sale Distribution")
plt.xlabel("Total Sale")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv(r"E:\Downloads\sales_data.csv")
print(df.head(15))
print()

print("Checking for missing values")
print(df.isna().sum())

print("Convert the Date column to proper datetime format.")

df["Date"]=pd.to_datetime(df["Date"], errors="coerce")
df["Year"]=df["Date"].dt.year
df["Month"]=df["Date"].dt.month
df["Revenue_per_Unit"]=df["Revenue"]/df["Units_Sold"]
print(df.head())
print()

print("Top 3 products")
print(df.groupby("Product")["Revenue"].sum().sort_values(ascending=False).head(3))
print()

print("Avg monthly sales per region")
print(df.groupby(["Region","Month"])["Units_Sold"].mean())
print()

print("get total revenue and average units sold per product.")
print(df.groupby("Product").agg({"Revenue":"sum","Units_Sold":"mean"}))
print()

print("Overall mean, median, and standard deviation of Revenue.")
rev=df["Revenue"].to_numpy()
print("Mean:",np.mean(rev))
print("Meadian:",np.median(rev))
print("SD:",np.std(rev))
rev_perunit=df["Revenue_per_Unit"].to_numpy()
print(rev_perunit)
print()

region_rev=df.groupby("Region")["Revenue"].sum()
plt.bar(region_rev.index,region_rev.values)
plt.title("Total Revenue per Region")
plt.show()

monthly_rev = df.groupby("Month")["Revenue"].sum()
plt.plot(monthly_rev.index, monthly_rev.values)
plt.title("Monthly Revenue Trend")
plt.show()

plt.scatter(df["Units_Sold"],df["Revenue"])
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

Employee_ID=np.arange(1,501)
Department=np.random.choice(["HR", "IT", "Finance", "Sales"],size=500)
Salary=np.random.randint(10000, 50000, size=500)
Experience_Years=np.random.randint(1,20, size=500)
Performance_Score=np.random.randint(1,5,size=500)

df=pd.DataFrame({"Employee_ID":Employee_ID,
                 "Department":Department,
                 "Salary":Salary,
                 "Experience_Years":Experience_Years,
                 "Performance_Score":Performance_Score})

missing_salary=np.random.choice(df.index,size=50, replace=False)
missing_exp=np.random.choice(df.index,size=50,replace=False)

df.loc[missing_salary,"Salary"]=np.nan
df.loc[missing_exp,"Experience_Years"]=np.nan

print(df.isna().sum())
print()

df["Salary"]=df["Salary"].fillna(df.groupby("Department")["Salary"].transform("median"))
df["Experience_Years"]=df["Experience_Years"].fillna(df["Experience_Years"].mean())

avg_salary=df.groupby("Department")["Salary"].mean()
print(avg_salary)
print()

avg_perf=df.groupby("Department")["Performance_Score"].mean()
print(avg_perf)
print()

exp=df.groupby("Department")["Experience_Years"].mean()
print(exp)
print()

plt.bar(avg_salary.index, avg_salary.values)
plt.title("average salary per department.")
plt.show()

plt.scatter(df["Experience_Years"],df["Salary"],c=df["Performance_Score"])
```
