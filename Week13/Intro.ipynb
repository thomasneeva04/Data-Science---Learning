#### Prove (AB)T=BTAT

```python
import numpy as np
A=np.array([[2,4,6],
            [8,10,12]])
B=np.array([[1,3,],
            [5,7],
             [9,11]])
left=(A@B).T
right=B.T @ A.T

print("Left: ",left)
print("Right: ",right)
```

#### FUNCTIONS

```python
def f(x):
    return 3*x+2

inputs = [0,1,2,3,4]
outputs= []

for x in inputs:
    y=f(x)
    outputs.append(y)

print("Input: ",inputs)
print("Output: ",outputs)
```

#### Q: If f(x)=x+10, find f(7)

```python
def f(x):
    return x+10

inputs = [7]
outputs = []

for x in inputs:
    y=f(x)
    outputs.append(y)

print("Output: ",outputs)
```

#### Checking many-to-one idea using values(Scratch)

```python
def f(x):
    return x**2

inputs=[-2,2]
outputs=[]

for x in inputs:
    outputs.append((f(x)))

print("Input: ",inputs)
print("Output : ",outputs)

if outputs[0]== outputs[1] and inputs[0]!= inputs[1]:
    print("This fn is many to one for this inputs")
else:
    print("This fn is not many to one for this inputs")
```
