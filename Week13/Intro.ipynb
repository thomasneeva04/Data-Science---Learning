#### Prove (AB)T=BTAT

```python
import numpy as np
A=np.array([[2,4,6],
            [8,10,12]])
B=np.array([[1,3,],
            [5,7],
             [9,11]])
left=(A@B).T
right=B.T @ A.T

print("Left: ",left)
print("Right: ",right)
```
