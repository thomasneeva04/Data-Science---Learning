```python
#Prime number using while
i=2
num=int(input("Enter a number: "))
limit=int(num/2)
if num>1:
    while i<=limit:
        if num%i==0:
            print(f"{num} is not a prime number")
            break
        i=i+1
    else:
        print(f"{num} is a prime number")
else:
    print("Entered number is 1 or lesser than 1")
```

```python
#prime number using for
num=int(input("Enter a number"))
i=2
limit=int(num/2)
for i in range(2,limit):
    if num%i==0:
        print("Not a prime number")
        break
else:
    print("Prime number")
```

```python
#Armstrong number
num=int(input("Enter a number: "))
temp=num
sum=0
n=len(str(num))
while temp>0:
    digits=temp%10
    sum+=digits**n
    temp//=10
if sum==num:
    print("Armstrong number")
else:
    print("Not an armstrong number")
    
```

```python
#array in descending order without using sort
num=int(input("Enter the number of elements: "))
list1=[]
for i in range(num):
    values=int(input("Enter the values: "))
    list1.append(values)

for i in range(num):
    for j in range(i+1,num):
        if list1[i]<list1[j]:
            list1[i],list1[j]=list1[j],list1[i]
print(list1)
```

```python
#palindrome without using slicing
a=input("Enter a word: ").upper()
length=len(a)
is_Palindrome=True

for i in range(length//2):
    if a[i]!=a[length-i-1]:
        is_Palindrome=False
        break
if is_Palindrome:
    print(f"{a} is a palindrome")
else:
    print(f"{a} is not a palindrome")
```
