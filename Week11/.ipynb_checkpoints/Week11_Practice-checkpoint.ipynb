### SIMULATION:
The process of repeating a random experiment many times and counting outcomes is called a simulation.

```python
import random
balls = ["Red","Red","Red","Red",
         "Blue","Blue","Blue",
         "Yellow","Yellow","Yellow"]
trials=1000
red_count=0
for _ in range(trials):   #The _ is used when: We don’t care about the loop variable. We just want the loop to repeat something a number of times.
    picked=random.choice(balls)
    if picked=="Red":
        red_count+=1
prob=red_count/trials
prob
```

Simulation (also called experimental probability) is a method where we estimate probability by repeating a random process many times and observing the outcomes. Instead of using formulas, we let Python randomly select values and count how often an event occurs. As the number of trials increases, the simulated probability gets closer to the theoretical probability. For example, when simulating random draws from a list of colored balls, the estimated probability of selecting a red ball was close to the theoretical 40%. Simulation helps validate probability results and is especially useful when manual calculations are complex or when working with real-world uncertainty.

#### Binominal Distribution: 
The binomial distribution is used when an experiment has only two possible outcomes, such as success or failure, and is repeated a fixed number of times with the same probability of success. It is defined by two parameters: n (number of trials) and p (success probability). The expected value is calculated as n × p, and the variance is n × p × (1 − p). Using Python, we simulated binomial outcomes with np.random.binomial() and analyzed the results using value_counts() to observe how simulated probabilities align with theoretical expectations. This helps visualize real-world uncertainty, especially in cases like conversion rates, email opens, or defect detection.

```python
import numpy as np

# parameters
n = 20   # number of trials
p = 0.3  # probability of success

# simulate 10,000 experiments
trials = 10000
results = np.random.binomial(n, p, trials)

results[:10]  # show first 10 outcomes
```

```python
import matplotlib.pyplot as plt
import pandas as pd

pd.Series(results).value_counts(normalize=True).sort_index().plot(kind='bar')
plt.title("Binomial Distribution: n=20, p=0.3")
plt.xlabel("Number of Clicks")
plt.ylabel("Probability")
plt.show()
```

Q. A marketing team observes that:
A website button has a conversion rate = 12%
On a given day, 200 visitors interact with it

We want to simulate:
👉 How many users will click the button?

```python
import numpy as np
import pandas as pd

n=200
p=0.12
trials=10000

result=np.random.binomial(n,p,trials)
pd.Series(result).value_counts(normalize=True).sort_values(ascending=False).head()
```

When I simulated the user click behavior using a binomial distribution, the most common outcome was around 24 clicks out of 200, which matches the theoretical expected value of 24. The distribution also showed values close to 24 occurring frequently, and extreme results appearing less often. This confirms the binomial behavior.

#### Z-Scores

```python
import numpy as np

# sample data (simulating normally distributed values)
data = np.random.normal(50, 10, 1000)

# calculate z scores
z_scores = (data - np.mean(data)) / np.std(data)

# show first 10 values
np.round(z_scores[:10], 2)
```

#### Q.Use binomial distribution to find the probability of exactly 5 purchases out of 10 trials with a success probability of 0.6.

```python
# Using PMF Formula
import numpy as np
from math import comb

n=10
p=0.6
k=5

prob_binomial= comb(n, k) * (p**k) * ((1-p)**(n-k))
np.round(prob_binomial,4)
```

```python
from scipy.stats import binom

n = 10  # number of trials
p = 0.6 # probability of success
k = 5   # exact number of successes

prob_binomial = binom.pmf(k, n, p)
prob_binomial.round(4)
```

```python
# Using Simulation

result=np.random.binomial(10,0.6,10000).tolist().count(5)/10000
result
```
