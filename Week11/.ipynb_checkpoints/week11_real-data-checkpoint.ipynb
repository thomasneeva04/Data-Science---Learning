```python
import pandas as pd
df=pd.read_csv('customer_data.csv')
df.head()
```

#### Finding mean/median/mode for Age

```python
#MEAN
#using function
mean_age=df['Age'].mean().astype(int)
#using Python
ages=list(df['Age'])
length=len(ages)
total=sum(ages)
M=total/length




#Median
#using Python(from scratch)
sorted_ages=sorted(ages)
n=len(sorted_ages)
if n%2==1:
    med_manual=sorted_ages[n//2]
else:
    med_manual=(sorted_ages[n//2-1]+sorted_ages[n//2])/2

#using Function
median_age=df['Age'].median()





#MODE(using frunction)
mode_age=df['Age'].mode()[0]

#using Python(from scratch)
from collections import Counter     #Counter is a built-in Python tool that counts how many times each value appears in a list.
frequency=Counter(ages)
mode_manual=frequency.most_common(1)[0][0]

print(M,mean_age)
print(med_manual, median_age)
print(mode_age,mode_manual)
```

So from the above data we understand that, the average age is 44. And the median age is showing as 45 which basically means slightly higher aged people's datas are present in the dataset And since the mode is 50, we can analyse them as most people has age 50 in the dataset. 

### Variance


```python
#using functions

age_var=df['Age'].var()
age_var

#using Python(scratch)
mean_age=sum(ages)/len(ages)
squared_diff=[(age-mean_age)**2 for age in ages]
var=sum(squared_diff)/(len(ages)-1)

print(age_var, var)
```

### Standard Deviation


```python
#using functions
age_std=df['Age'].std()


#using Python(scratch)
import math
age_std_manual=math.sqrt(var)

age_std,age_std_manual
```

So from this we understand that, ages in the dataset vary about 15 years from the avg age. This means, not a lot of  ages are clustered aroung avg age. 

### InterQuartile Range

```python
import numpy as np

Q1=np.percentile(df['Age'],25)
Q2=np.percentile(df['Age'],75)
IQR=Q2-Q1

print(Q1,Q2,IQR)
```

This means that middle 50% of ages falls between 57 and 32. 

### Skew

```python
df['Age'].skew().round(2)
```

The skew values is -0.11 which is very near to zero. This means the distribution is symmetric. The values are fairly balanced around the center and there is no tilt towrds left/right. 

### Kurtosis

```python
df['Age'].kurt().round(2)
```

The kurtosis value is -1.11 which is below 0. This means the dist is flatter and has values which are widely spread out comparing to a normal distribution. Also less outliers are present in this data. 

### Visualization

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(7,4))
plt.hist(df['Age'],bins=10, edgecolor='black')
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.plot
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.boxplot(df['Age'],vert=False)
plt.xlabel("Age")
plt.title("Boxplot of Age")
plt.show()
```

The median is roughly in the center of the box, so the data is fairly balanced. The middle 50% of values aren’t too close or too spread out — they sit in a moderate range. The whiskers are also moderate, and there are no outliers. So overall, the data isn’t too tight or too scattered — it’s somewhere in the middle.

```python
df['Gender'].value_counts().plot(kind='bar')
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Gender Distribution")
plt.show()
```

The Male and Female counts look almost the same, with Male just slightly higher. There isn’t a big difference between the two main groups. However, the “Other” category is noticeably lower and appears only in a small portion of the dataset.

```python
df.groupby('Gender')['Income'].mean().plot(kind='bar')
plt.xticks(rotation=0)
plt.xlabel("Gender")
plt.ylabel("Income")
plt.title("Avg income by Gender")
plt.show()
```

The frequency chart showed that the “Other” category has a very small representation. However, the value-based bar chart shows that their average income is close to Male and Female. This means that although the category size is small, the income distribution for those individuals is similar or slightly higher, which explains why the average remains closer to the other groups.

```python
plt.figure(figsize=(8,4))
plt.scatter(df['Purchase_Count'], df['Wait_Time_Minutes'])
plt.xlabel("Wait Time")
plt.ylabel("Purchase Count")
plt.title("Purchse Count vs Wait Time Minutes")
plt.show()
```

There isn’t a pattern here. Waiting longer doesn’t seem to increase or decrease purchases. People who wait 2 minutes and people who wait 10 minutes both show similar purchase behavior, so there's no visible link between waiting time and how much someone buys.

```python
df[['Wait_Time_Minutes', 'Purchase_Count']].corr()
```

I checked the relationship between wait time and purchase count using both a scatter plot and correlation. The scatter plot showed no visual pattern, and the correlation value was 0.0115, which is almost zero. This means there’s no linear relationship — waiting longer doesn’t impact how many items customers buy.
